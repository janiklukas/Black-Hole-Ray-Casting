```julia
#Choose an ODE solver (euler, RK2 or RK4)
solver::String = "RK4";
```

```julia
#camera params
r::Float64 = 2.5
θ::Float64 = π/2 - 0.07
dist::Float64 = 0.1
fov::Float64 = π/2

#numerical params
Δt::Float64 = 0.01
t_max::Float64 = 3.0
ρ::Float64 = 1.25
r_max::Float64 = 5.0
iters::Int64 = 3

#image params
px::Int64 = 300
size::Int64 = 300
star_prob::Float64 = 0.02

camera = Camera(r, θ, dist, fov)
generate_image(camera, px, Δt, t_max, ρ, r_max, iters, size; star_prob)
```

```julia
#Animation

Δt_array = collect(range(start = 0.3, stop = 0.01, length = 100))

fps::Int64 = 10
gif_length::Float64 = 5.0
px::Int64 = 100

animate_timesteps(camera, px, solver, Δt_array, t_max, ρ, r_max, iters, size, fps, gif_length)
```
